# 숫자 블록
- `유형` 연습문제



- `문제 설명` 


그렙시에는 0으로 된 도로에 숫자 블록을 설치하기로 하였습니다. 숫자 블록의 규칙은 다음과 같습니다.


블록의 번호가 n 일 때, 가장 처음 블록은 n * 2번째 위치에 설치합니다. 그다음은 n * 3, 그다음은 n * 4, ...로 진행합니다.만약 기존에 블록이 깔려있는 자리라면 그 블록을빼고 새로운 블록으로 집어넣습니다.


예를 들어 1번 블록은 2,3,4,5, ... 인 위치에 우선 설치합니다. 그다음 2번 블록은 4,6,8,10, ... 인 위치에 설치하고, 3번 블록은 6,9,12... 인 위치에 설치합니다.


이렇게 3번 블록까지 설치하고 나면 첫 10개의 블록은 0, 1, 1, 2, 1, 3, 1, 2, 3, 2이됩니다.


그렙시는 길이가 1,000,000,000인 도로에 1번 블록부터 시작하여 10,000,000번 블록까지 위의 규칙으로 모두 놓았습니다.


그렙시의 시장님은 특정 구간의 어떤 블록이 깔려 있는지 알고 싶습니다.


구간을 나타내는 두 수 begin, end 가 매개변수로 주어 질 때, 그 구간에 깔려 있는 블록의 숫자 배열(리스트)을 return하는 solution 함수를 완성해 주세요.


- `제한 조건`
    - begin, end 는 1 이상 1,000,000,000이하의 자연수 이고, begin는 항상 end보다 작습니다.
    - end - begin 의 값은 항상 10,000을 넘지 않습니다.


- `입력`
    - ex1) 1, 10


- `출력`
    - ex1) [0, 1, 1, 2, 1, 3, 1, 4, 3, 5]
    - 아래와 같이 블럭이 깔리게 됩니다.

```python
# TRY 1) 정확성 통과, 효율성 미통과

def divisor(x):
    for i in range(2, x+1):
        if x % i == 0:
            return x // i 
    return 0
            
# begin, end을 포함하는 리스트 생성 후
# 두 구간 사이를 순회하면서 2번째로 큰 약수를 매회 구해서 리스트에 삽입
def solution(begin, end):
    lst = [1] * (end-begin+1)
    for idx, num in enumerate(range(begin, end+1)):
        lst[idx] = divisor(num)
    return lst
```

```python
print(solution(1, 10)) # [0, 1, 1, 2, 1, 3, 1, 4, 3, 5]
print(solution(50000000, 50000010)) # [10000000, 2631579, 3846154, 101833, 8333334, 4545455, 7142858, 1, 6250001, 2941177, 8333335]
```

```python
# SOL 1)

# 1을 제외한 가장 작은 약수로 나누어준 몫이 10_000_000 이하인 값을 구하는 함수
# 나누어지는 수가 1인 경우 및 위의 경우에 해당되지 않는 경우 예외 처리 
def divisor(x):
    if x == 1:
        return 0
    else:
        for i in range(2, int(x**0.5)+1):
            if x % i == 0 and x // i <= 10_000_000:
                return x // i
        return 1
    
def solution(begin, end):
    
    # begin, end을 포함하는 리스트 생성
    lst = [1] * (end-begin+1)
    
    # begin과 end 구간을 순회하며 숫자를 받아옴
    for idx, num in enumerate(range(begin, end+1)):
        lst[idx] = divisor(num)
    return lst
```

```python
print(solution(1, 10)) # [0, 1, 1, 2, 1, 3, 1, 4, 3, 5]
print(solution(50000000, 50000010)) # [10000000, 2631579, 3846154, 101833, 8333334, 4545455, 7142858, 1, 6250001, 2941177, 8333335]
```

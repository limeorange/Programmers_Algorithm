- `` 2018 KAKAO BLIND RECRUITMENT


- ` ` 


         .            .         .


    1.     n   ,   ""(" ")  ""("#")    .
    2.         .  " 1" " 2" .  1   2        .  1  2      .
    3. " 1" " 2"     .
    4.        1,   0       .



      ,       .


- ``
    - ex1) 5, [9, 20, 28, 18, 11], [30, 1, 21, 17, 28]
    - ex2) 6, [46, 33, 33, 22, 31, 50], [27, 56, 19, 14, 14, 10]


- ``
    - ex1) ["#####","# # #", "### #", "# ##", "#####"]
    - ex2) ["######", "### #", "## ##", " #### ", " #####", "### # "]

```python
# SOL 1)

def solution(n, arr1, arr2):
    b1 = []
    for i in arr1:
        i = format(i, 'b')
        while len(i) < n:
            i = '0'+i
        b1.append(i)

    b2 = []
    for i in arr2:
        i = format(i, 'b')
        while len(i) < n:
            i = '0'+i
        b2.append(i)

    code = []
    for i, j in zip(b1, b2):
        temp = ''
        for ii, jj in zip(i, j):
            if ii == jj == '0':
                temp += '0'
            else:
                temp += '1'
        code.append(temp)

    result = []
    for i in code:
        i = i.replace('1', '#').replace('0', ' ')
        result.append(i)

    return result
```

```python
print(solution(5, [9, 20, 28, 18, 11], [30, 1, 21, 17, 28]))
print(solution(6, [46, 33, 33, 22, 31, 50], [27, 56, 19, 14, 14, 10]))
```
